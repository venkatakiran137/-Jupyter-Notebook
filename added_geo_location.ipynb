# requesting for the data by using beautiful soup

```python
postel_codes= requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
df=pd.read_html(postel_codes.content)[0]
```

```python
df.head()
```

**removing not assigned values from borough**

```python
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
```

```python
df
```

**grouping the postal code for More than one neighborhood can exist in one postal code area.**

```python
df1=df.groupby("Postal Code").agg(lambda x:','.join(set(x)))
```

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']
```
